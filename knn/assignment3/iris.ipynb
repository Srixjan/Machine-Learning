```python
import pandas as pd
# import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_score, recall_score, accuracy_score, f1_score, confusion_matrix, classification_report

iris = pd.read_csv("iris.csv")
```

```python
iris.head()           # first 5 rows — see structure
iris.shape            # how many rows and columns
iris.info()           # data types, null counts
iris.describe()       # mean, min, max, std for each column
iris.isnull().sum()   # check for missing values
iris["Species"].value_counts()  # check class balance
```

```python
X = iris.drop(["Species"], axis=1)
y = iris['Species']

X_train, X_test, y_train, y_test = train_test_split (
    X, y,
    test_size = 0.50,
    random_state = 42
)

X_test_full = X
y_test_full = y

print(f"Train: {len(X_train)} samples | Test (full): {len(X_test_full)} samples")
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_full_scaled = scaler.transform(X_test_full)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_scaled, y_train) 

lR = LogisticRegression(max_iter=200)
lR.fit(X_train_scaled, y_train)

nb = GaussianNB()
nb.fit(X_train_scaled, y_train)

y_knn = knn.predict(X_test_full_scaled)
y_lR = lR.predict(X_test_full_scaled)
y_nb = nb.predict(X_test_full_scaled)
```

```python
models = {
    "KNN" : y_knn,
    "Logistic Regression" : y_lR,
    "Naives Bayes" : y_nb
}
for name, y_pred in models.items():
    acc = accuracy_score(y_test_full, y_pred)
    print(f"{name:25s} → {acc*100:.1f}%")
```
