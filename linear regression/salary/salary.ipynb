```python
import pandas as pd
import seaborn as sns
```

```python
salary = pd.read_csv("Salary_dataset.csv")
```

```python
salary.shape
salary.describe()
```

```python
# Dataset has no null values.
salary.isnull().sum()
salary.drop(["Unnamed: 0"], axis=1, inplace=True)
```

```python
# As we can see there is gradual growth in salary, in-correspondence to years.
sns.scatterplot(x=salary['YearsExperience'], y=salary['Salary'])
```

```python
# x contains only Years of Expierence, it is an independent variable.
# y contains Salary, as its a dependent variable, as it
x = salary.drop(columns=['Salary'])
y = salary['Salary']

#Train test split is used to split the datasets into tests and training, in the 80-20 rule.
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split (
    x, y, test_size=0.20, random_state=45
)
```

```python
#Training the model.
from sklearn.linear_model import LinearRegression 

model = LinearRegression()
model.fit(x_train, y_train)
```

```python
# Running the model, on x_test and getting predicted values for y_pred.
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print("R2 score: ",r2*100,"%")
```
