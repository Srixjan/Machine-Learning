```python
import pandas as pd
import seaborn as sns

store = pd.read_csv("store.csv")
test = pd.read_csv("test.csv")
train = pd.read_csv("train.csv" ,low_memory=False)
```

```python
print(train.shape, test.shape, store.shape)
```

```python
train = train.merge(store, on='Store', how='left')
test = test.merge(store, on='Store', how='left')
```

## 'train' dataset has been cleaned, pre processed, and encoded for model training.

```python
train.isnull().sum()
```

```python
train['CompetitionDistance'] = train['CompetitionDistance'].fillna(train['CompetitionDistance'].median())

train = train.drop(columns=['PromoInterval'])

train['CompetitionOpenSinceMonth'] = train['CompetitionOpenSinceMonth'].fillna(0)

train['CompetitionOpenSinceYear'] = train['CompetitionOpenSinceYear'].fillna(0)

train['Promo2SinceWeek'] = train['Promo2SinceWeek'].fillna(0)

train['Promo2SinceYear'] = train['Promo2SinceYear'].fillna(0)
```

```python
train.head(5)
```

```python
cols = ['StoreType', 'Assortment', 'StateHoliday']
train = pd.get_dummies(train, columns=cols, drop_first=True)
```

```python
train['Date'] = pd.to_datetime(train['Date'])

for df in [train]:
    df['Year'] = df['Date'].dt.year
    df['Month'] = df['Date'].dt.month
    df['Week'] = df['Date'].dt.isocalendar().week

train.drop(columns=['Date'], inplace=True)
```

```python
train.head(10)
```

## 'test' data is going to be pre-processed, cleaned, and enncoded for future model running.

```python
print(test.shape)
print(test.describe())
print(test.head(5))
```

```python
test['CompetitionDistance'] = pd.to_numeric(test['CompetitionDistance'], errors='coerce')
median_distance = train['CompetitionDistance'].median()
```

```python
test['Open'] = test['Open'].fillna(1)

test['CompetitionDistance'] = test['CompetitionDistance'].fillna(median_distance)

test['CompetitionOpenSinceMonth'] = test['CompetitionOpenSinceMonth'].fillna(0)

test['CompetitionOpenSinceYear'] = test['CompetitionOpenSinceYear'].fillna(0)

test['Promo2SinceWeek'] = test['Promo2SinceWeek'].fillna(0)

test['Promo2SinceYear'] = test['Promo2SinceYear'].fillna(0)

test.drop(columns=['PromoInterval'], inplace=True)

train = train.drop(columns=['Customers'])
```

```python
print(test['StateHoliday'].unique(), test['StoreType'].unique(), test['Assortment'].unique())
```

```python
cols = ['StateHoliday', 'StoreType', 'Assortment']
test = pd.get_dummies(test, columns=cols, drop_first=True)
```

```python
test['Date'] = pd.to_datetime(test['Date'])

for df in [test]:
    df['Year'] = df['Date'].dt.year
    df['Month'] = df['Date'].dt.month
    df['Week'] = df['Date'].dt.isocalendar().week

test.drop(columns=['Date'], inplace=True)
```

##  Using "Train" dataset to do Linear Regression model.

```python
from sklearn.model_selection import train_test_split 
X  = train.drop(columns=['Sales'])
Y = train['Sales']

X_train, X_test, Y_train, Y_test = train_test_split(
    X,Y, test_size=0.20, random_state=42
)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from sklearn.metrics import mean_absolute_percentage_error
import numpy as np

Y_pred = model.predict(X_test)

print("R2 Score:", r2_score(Y_test, Y_pred)*100)
print("MAE:", mean_absolute_error(Y_test, Y_pred))
print("RMSE", np.sqrt(mean_squared_error(Y_test, Y_pred)))
print(mean_absolute_percentage_error(Y_test,Y_pred))
```

```python
# Evalutate 
from sklearn.metrics import r2_score
r2 = r2_score(Y_test, Y_pred)
print("r-squared: ",r2*100,"%")

n = X_test.shape[0]
p = X_test.shape[1]

adjusted_r2 = 1 - ((1-r2) * (n-1)/n-p-1)
print("adjusted r^2 value: ", adjusted_r2)
```
